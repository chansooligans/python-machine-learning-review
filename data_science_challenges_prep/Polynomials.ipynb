# Non-Linear Regressions

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.names
df = pd.read_csv('data/boston.csv')
df.head()
```

## Scatter Plots

```python
g = sns.pairplot(df)
```

```python
sns.pairplot(df[['lowstatus','medv','access']],hue='access')
```

## Simple Linear Regression

```python
X = df[['distance']]
y = df[['nox']]
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
```

```python
print("R^2: {}".format(model.score(X,y)))
rmse = np.sqrt(mean_squared_error(y,y_pred))
print("RMSE: {}".format(rmse))
```

```python
plt.scatter(X, y, s=10)
plt.plot(X, y_pred, color='r')
plt.show()
```

## Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures

polynomial_features = PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)
```

```python
print("R^2: {}".format(model.score(x_poly, y)))
rmse = np.sqrt(mean_squared_error(y, y_poly_pred))
print("RMSE: {}".format(rmse))
```

```python
# Sort Predicted Values on X1
y_poly_pred = pd.DataFrame(y_poly_pred,columns=['y_poly_pred'])
df2 = pd.concat([X, y_poly_pred], axis=1)
df2 = df2.sort_values('distance')

# Plot
plt.scatter(X, y, s=10)
plt.plot(df2.distance, df2.y_poly_pred, color='m')
plt.show()
```

#### CV Model

```python
from sklearn.base import BaseEstimator

class PolynomialRegression(BaseEstimator):
    def __init__(self, deg=None):
        self.deg = deg
    
    def fit(self, X, y, deg=None):
        polynomial_features = PolynomialFeatures(self.deg)
        x_poly = polynomial_features.fit_transform(X)
        self.model = LinearRegression(fit_intercept=False)
        self.model.fit(x_poly, y)
    
    def predict(self, x):
        polynomial_features = PolynomialFeatures(degree=self.deg)
        x_poly = polynomial_features.fit_transform(x)
        return self.model.predict(x_poly)
    
    @property
    def coef_(self):
        return self.model.coef_
```
